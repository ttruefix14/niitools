```python
import pandas as pd
import re
```

```python
xlsx = pd.ExcelFile(r"../defaults/p10-2024.xlsx")
```

```python
df = pd.read_excel(xlsx, "Общий")
```

```python
df
```

```python
df = df.loc[df["Condition"] != "True"]
```

```python
result_dict = {}
for i, row in df.iterrows():
    
    if not result_dict.get(row.Layer):
        result_dict[row.Layer] = {}

    conditions = row.Condition.split("and")
    regexpin = r'.+\'(.+)\'.+(\[.+\]).*'
    regexpeq = r'.+\'(.+)\'.+=.+(\d).*'
    regexprange = r'.+\'(.+)\'.+range\( *(\d.+)\).*'

    conds = "{"
    for c in conditions:
        if "range" in c:
            result = re.sub(regexprange, r'"\1": [\2]', c).replace(" + ", "")
            print(row.Layer, row.Name, result)
        elif "in" in c:
            result = re.sub(regexpin, r'"\1": \2', c)
            print(row.Layer, row.Name, result)
        elif "=" in c:
            result = re.sub(regexpeq, r'"\1": [\2]', c)
            print(row.Layer, row.Name, result)
        conds += result + ", "
    conds += "}"

    result_dict[row.Layer][row.Name] = eval(conds)
```
